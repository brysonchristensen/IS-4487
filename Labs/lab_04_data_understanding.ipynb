## Part 2: Inspecting, Cleaning & Removing Outliers

### Instructions:
- Identify missing values.
- Check data types are correct(e.g., dates, numeric columns).
- Check for outliers key numeric variables like `price`, `sqft`, `beds`, or `baths`.

### Think about:
- Are there missing values?
- Do the data types look appropriate?
- Do any numeric variables contain extreme or unusual values?



```python
# Check for missing values
df.isnull().sum()
```

Yes, there are missing values in several columns, including county, beds, baths, sqft, address, lat, lon, title, descr, and details. The columns with the most missing values are address, lat, lon, descr, and details.

```python
# Basic summary statistics
df[['price', 'beds', 'baths', 'sqft']].describe()
```

```python
# Check data types
df.dtypes
```

The data types generally look appropriate. post_id, nhood, city, county, address, title, descr, and details are objects (strings), which makes sense. date and year are integers, which is fine, although date might be more useful as a datetime object for time-series analysis. price, beds, baths, sqft, lat, and lon are numeric (int64 or float64), which is also appropriate. room_in_apt is an integer, representing a binary variable, which is correct.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot of price
sns.boxplot(x=df['price'])
plt.title("Boxplot of Rental Price")
plt.show()
```

Based on the descriptive statistics and the boxplot of price, it appears that there are some extreme or unusual values in the price column. The maximum price is significantly higher than the 75th percentile, and the boxplot shows several outliers.

```python
# Remove price outliers (1stâ€“99th percentile)
q_low = df['price'].quantile(0.01)
q_high = df['price'].quantile(0.99)
df = df[(df['price'] >= q_low) & (df['price'] <= q_high)]
```

### ðŸ”§ Try It Yourself â€“ Part 2

1. Use `.describe()` and a boxplot to check for outliers in **square footage (`sqft`)**.

2. What patterns or issues do you see with square footage values? Is there anything unusual?


```python
# Check for outliers in sqft using describe()
print(df['sqft'].describe())

# Boxplot of sqft
sns.boxplot(x=df['sqft'])
plt.title("Boxplot of Square Footage")
plt.show()
```

ðŸ”§ Add comment here: Based on the .describe() output and the boxplot for sqft, there are clearly extreme outliers. The maximum value (900,000 sqft) is unrealistically large for a rental property and significantly skews the mean and standard deviation. This suggests data entry errors or incorrect units for some listings. These outliers should likely be investigated or removed before further analysis.

## Part 3: Basic Exploration

Use `groupby` and `value_counts` to summarize trends across neighborhoods and cities.


```python
# Average price by neighborhood
df.groupby('nhood')['price'].mean().sort_values(ascending=False).head(10)
```

```python
# Top cities by count
df['city'].value_counts().head(10)
```

### ðŸ”§ Try It Yourself â€“ Part 3

Explore the data by performing both of the following:

1. Group the listings by `year` and calculate the average price for each year.
2. Use `.value_counts()` on the `room_in_apt` column to see how common room rentals are.

3. Add a short comment or markdown cell describing any trends or insights you found.


```python
# Average price by year
print("Average price by year:")
print(df.groupby('year')['price'].mean())

# Count of room_in_apt values
print("\nValue counts for room_in_apt:")
print(df['room_in_apt'].value_counts())
```

Room Rentals (room_in_apt): The value_counts() for room_in_apt shows that listings for entire apartments (0) are significantly more common than listings for single rooms within an apartment (1).
